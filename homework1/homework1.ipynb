```python
import numpy as np
import matplotlib.pyplot as plt


def En(r):
    return (-1.436 / r) + (5.86e-6) / (r ** 9)

def Ea(r):
    return (-1.436 / r) 

def Er(r):
    return (5.86e-6) / (r ** 9)

r = np.linspace(0.01, 1, 200)

plt.plot(r, En(r), label='$E_N$')
plt.plot(r, Ea(r), label='$E_A$', linestyle='--')
plt.plot(r, Er(r), label='$E_R$', linestyle='--')

plt.title('Energy (eV) vs r (nm)')
plt.xlim(0.01, 1)
plt.ylim(-7, 7)

plt.xlabel('r (nm)')
plt.ylabel('Energy (eV)')
plt.legend()
plt.savefig('2a.png', dpi=300)
```

```python
# put horizontal line at E_N miniumum

x = r
y = En(r)

minY = np.min(y)
minX = x[np.argmin(y)]

plt.plot(x, y, label='$E_N$')

plt.title('Energy (eV) vs r (nm)')
plt.xlim(0.01, 1)
plt.ylim(-5, 3)

plt.xlabel('r (nm)')
plt.ylabel('Energy (eV)')

plt.axvline(x=minX, color='r', linestyle='--') # add vertical line

plt.plot([], [], 'r--', label=f'Equilibrium distance {minX:.4f} nm')
plt.legend()
plt.savefig('2b.png', dpi=300)
```
